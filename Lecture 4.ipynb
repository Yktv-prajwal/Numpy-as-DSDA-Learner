```python
import numpy as np

marks=np.array([70,45,34,76,35])
print("Student Marks",marks)
```

```python
mean= np.mean(marks)
print("Average Marks=",mean)
```

```python
std=np.std(marks)

print("Standard Deviation :",round(std,2))
```

```python
z_score=(marks-mean)/std
print("Z Score =",z_score)
```

```python
for mark, z in zip(marks,z_score):
    print(f"Marks = {mark}, Z Score = {z:.2f}")
```

```python
siample_mean=np.mean(marks)
population_mean=50
sample_sd=np.std(marks,ddof=1)
n=len(marks)

t_score=(siample_mean-population_mean)/(sample_sd/np.sqrt(n))
print("T Score =",t_score)

# Degree Of Freedom

df= n - 1 
print("Degree Of Freedom = ",df)
```
