```python
import numpy as np
```

```python
from numpy import random
```

```python
x = random.randint(100)
print(x)
```

```python
x = random.rand()
print(x)
```

```python
x = random.randint(100,size=(3,5))
print(x)
```

```python
# Float Array
x = random.rand(5)
print(x)

#Create 2D Array
x = random.rand(3,5)
print(x)
```

```python
#Single Value
x = random.choice([3,4,5,6,7,8,9,10])
print(x)
```

```python
# Array of values 2d array
x = random.choice([1,2,3,4,5,6,7,8,9],size=(3,3))
print(x)
```

```python
colour = ["red","green","blue","lightgreen","pink"]
selected = np.random.choice(colour,size=3)
print(selected)
```

```python
#Shuffle in array
arr = np.array([21,22,23,24,25])
np.random.shuffle(arr)

#permuation
new_arr = np.random.permutation(arr)
print(new_arr)
```

```python
#Seed()
np.random.seed(42)
x = np.random.randint(1,100)
print(x)
```
