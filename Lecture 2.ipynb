```python
import numpy as np
data = np.array([12,13,14,15,16,17])
print(data[3])
print(data[-2])

#SLicing
print(data[1:4])
print(data[:4])
print(data[2:])
print(data[:])
```

```python
data_2d = np.array([
    [21,22,23],
    [24,25,26],
    [27,28,29]
])

print(data_2d)
print(data_2d.ndim)
print(data_2d.shape)

print(data_2d[1,1])
print(data_2d[0,1])

#access 2nd column

print(data_2d[:,1])
print(data_2d[0,:])
```

```python
arr_3d = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(arr_3d)

#Ellipsis (...)
print(arr_3d[...,0])
```

```python
print(arr_3d[...,1])
```

```python
print(arr_3d[1][0,1]) #remove single number
```

```python
print(arr_3d[0][1,:])
```

```python
print(arr_3d[1][1,:])
```
